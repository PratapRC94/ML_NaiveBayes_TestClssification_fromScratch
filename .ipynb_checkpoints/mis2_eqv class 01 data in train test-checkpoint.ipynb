```python
import re
import os
import sys
import math
import random
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
#%matplotlib inline
from matplotlib.pyplot import figure

import time
start = time.time()
#print('time when start : \n',start)

def read_files(filename):
    with open("pp1data/" + filename, "r",encoding='Latin-1') as f:
        data = f.read()
    return data


def pre_processing(text):
    
    # Convert string to lower case
    text1 = str(text).lower() 
    # Removing html tags and content inside it
    text1 = re.sub(r'<.*?>', '', text1)
    # Removing URLs
    text1 = re.sub(r'[a-z]*[:.]+\S+', '', text1)
    # Remove punctuations
    text1 = text1.replace("'","")
    text1 = re.sub(r'[^\w\s]',' ',text1)
    # Remove hyphen,underscore
    text1 = re.sub('-|_', ' ',text1)
    # Removing digits
    text1 = re.sub('\d+', "", text1)
    # Removing words with 2 letters and less
    text1 = re.sub(r'\b\w{1,2}\b', '',text1)
    # Remove extra space with ''
    text1 = re.sub("\s\s+"," ",text1)
    
    return text1


if __name__ == "__main__": 

    #source_data = read_files(sys.argv[1])
    source_data = read_files(input())
    am = source_data.lower().split("\n")
    am = list(string for string in am if len(string)>0)

    am_processed={}
    for i in range(len(am)):
        am_processed[i] = list([pre_processing(am[i].split('\t')[0]),am[i].split('\t')[1]])
    len(am_processed)


    lb_0={} #neg
    lb_1={} #pos
    for i in range(len(am)):
        if am_processed[i][1]=='0':
            lb_0[i] = am_processed[i]
        else:
            lb_1[i] = am_processed[i]
    m = [0,1]
    k=[2,3,4,5,6,7,8,9,10]
    avg_accuracy_parameter_wise = [] # list of average of accuracies over k_fold(2,3,4,5,6,7,8,9,10) over m(0,1)
    for p in m :
        avg_accuracy_for_k_fold = [] # list of average of accuracies over k_fold(2,3,4,5,6,7,8,9,10)
        for q in k:
            accuracy= [] # list of accuracy for each iteration of k_fold
            for j in range(q):
                
                #am_to_be_shuffled = list(am_processed.items())
                #random.shuffle(am_to_be_shuffled)
                #am_shuffled = dict(am_to_be_shuffled)
                
                
                lb_0_to_be_shuffled = list(lb_0.items())
                random.shuffle(lb_0_to_be_shuffled)
                lb_0_shuffled = dict(lb_0_to_be_shuffled)
                
                lb_1_to_be_shuffled = list(lb_1.items())
                random.shuffle(lb_1_to_be_shuffled)
                lb_1_shuffled = dict(lb_1_to_be_shuffled)
                
                '''
                #creating test data set from entire vocabulary - dictionary contains both text and label
                test = dict(list(am_shuffled.items())[j*int(len(am_shuffled)/q):(j*int(len(am_shuffled)/q))+int(len(am_shuffled)/q)])
                #creating train data set from entire vocabulary
                train = {}
                for i in set(am_shuffled.keys()) - set(test.keys()):
                    train[i] = am_shuffled[i]
                '''

                lb_1_test = dict(list(lb_1_shuffled.items())[j*int(len(lb_1_shuffled)/q):(j*int(len(lb_1_shuffled)/q))+int(len(lb_1_shuffled)/q)])
                lb_1_train = {}
                #lb_1_train = dict(set(lb_1.items()) - set(lb_1_test.items()))
                for i in set(lb_1_shuffled.keys()) - set(lb_1_test.keys()):
                    lb_1_train[i] = lb_1[i]

                #getting negative examples for testing data
                lb_0_test = dict(list(lb_0_shuffled.items())[j*int(len(lb_0_shuffled)/q):(j*int(len(lb_0_shuffled)/q))+int(len(lb_0_shuffled)/q)])
                #getting negative examples for training data
                lb_0_train = {}
                #lb_0_train = dict(set(lb_0.items()) - set(lb_0_test.items()))
                for i in set(lb_0_shuffled.keys()) - set(lb_0_test.keys()):
                    lb_0_train[i] = lb_0[i]

                
                '''
                #splitting the train and test data according to class label 0 and 1 with only text data in dictionary
                lb_1_test={} #pos
                lb_1_train={}
                lb_0_test={} #neg
                lb_0_train={}

                for i in (train.keys()):
                    if train[i][1]=='0':
                        lb_0_train[i] = train[i][0]
                    else:
                        lb_1_train[i] = train[i][0]
                #print(len(lb_1_train),len(lb_0_train))

                for i in (test.keys()):
                    if test[i][1]=='0':
                        lb_0_test[i] = test[i][0]
                    else:
                        lb_1_test[i] = test[i][0]

                #print(len(lb_1_test),len(lb_0_test))
                '''
                #****************************************************************************************************************

                #merging class 0/1 wise train and test data
                training_data = {**lb_1_train,**lb_0_train}
                test_data = {**lb_1_test,**lb_0_test}

                #token_1 =[]
                #token_0 =[]

                token_1 = []
                for sent in lb_1_train.values():
                    token_1.extend(sent[0].split())
                    #for word in sent.split():
                        #token_1.append(word)

                token_0 = []
                for sent in lb_0_train.values():
                    token_0.extend(sent[0].split())
                    #for word in sent.split():
                        #token_0.append(word)

                train_tokens = token_1+token_0

                #****************************************************************************************************************

                prob_1 = np.log(len(lb_1_train)/(len(lb_1_train) + len(lb_0_train)))
                prob_0 = np.log(len(lb_0_train)/(len(lb_1_train) + len(lb_0_train)))

                #prob_1 = len(lb_1_train)/(len(lb_1_train) + len(lb_0_train))
                #prob_0 = len(lb_0_train)/(len(lb_1_train) + len(lb_0_train))
                #print(prob_1,prob_0)
                #****************************************************************************************************************

                # for tokens in train_tokens but not in class 0
                train_tokens_not_0 = set(train_tokens) - set(token_0)
                train_tokens_not_0_dict = dict.fromkeys(list(train_tokens_not_0), 0)
                #dictionary of frequency of each token in class 0
                token_0_mle = dict(Counter(token_0))

                neg_counts = {**train_tokens_not_0_dict,**token_0_mle}

                #MAP estimate of tokens of class 0
                map_neg_prob ={}
                for key in neg_counts.keys():
                    map_neg_prob[key] = np.log((neg_counts[key]+p)/(sum(token_0_mle.values())+(p*len(set(train_tokens)))))
                    #map_neg_prob[key] = (neg_counts[key]+p)/(sum(token_0_mle.values())+(p*len(set(train_tokens))))

                # for tokens in train_tokens but not in class 1
                train_tokens_not_1 = set(train_tokens) - set(token_1)
                train_tokens_not_1_dict = dict.fromkeys(list(train_tokens_not_1), 0)
                #dictionary of frequency of each token in class 1
                token_1_mle = dict(Counter(token_1))

                pos_counts = {**train_tokens_not_1_dict,**token_1_mle}
                #MAP estimate of tokens of class 1
                map_pos_prob ={}
                for key in pos_counts.keys():
                    map_pos_prob[key] = np.log((pos_counts[key]+p)/(sum(token_1_mle.values())+(p*len(set(train_tokens)))))
                    #map_pos_prob[key] = (pos_counts[key]+p)/(sum(token_1_mle.values())+(p*len(set(train_tokens))))


                #****************************************************************************************************************
                ## class prediction on test data set

                final_class = {}
                for key,sent in test_data.items():  
                    #words = sent.split()
                    words = sent[0].split()
                    available_words = set(words).intersection(set(train_tokens)) #skip the words not available in training set

                    #MAP estimate of tokens from test data using the MAP of tokens from train data 
                    class_1 = 0
                    class_0 = 0

                    class_1 += prob_1
                    class_0 += prob_0

                    for word in available_words:
                        class_0 = class_0 + np.log(map_neg_prob[word])
                        class_1 = class_1 + np.log(map_pos_prob[word])
                        #class_0 = class_0 * map_neg_prob[word]
                        #class_1 = class_1 * map_pos_prob[word]
                    #class_0 = class_0 * prob_0
                    #class_1 = class_1 * prob_1

                    if class_1 > class_0:
                        final_class[key] = '1'
                    elif class_1 == class_0:
                        temp = random.randint(0, 1)
                        if temp == 1:
                            final_class[key] = '1'
                        else:
                            final_class[key] = '0'
                    else:
                        final_class[key] = '0'


                #****************************************************************************************************************
                actual_test_class = {}
                for key in test_data.keys():
                #for key in test.keys():
                    actual_test_class[key] = test_data[key][1]
                    
                #****************************************************************************************************************
                # Accuracy
                accuracy.append(len(set(final_class.items()).intersection(set(actual_test_class.items())))/len(actual_test_class))
                '''
                matched=0
                unmatched =0
                for i in actual_test_class.keys():
                    if actual_test_class[i]==final_class[i]:
                        matched=matched+1
                    else:
                        unmatched=unmatched+1
                accuracy = accuracy+[((matched)/(matched+unmatched))]
                '''

            #****************************************************************************************************************

            sum_of_accuracy = sum(accuracy)
            #for i in range(len(accuracy)):
                #sum_of_accuracy + = accuracy[i]
            avg_accuracy = np.mean(accuracy)
            #avg_accuracy = sum_of_accuracy/len(accuracy)
           # print(avg_accuracy)
            avg_accuracy_for_k_fold.append(avg_accuracy)

       # print(avg_accuracy_for_k_fold)
        avg_accuracy_parameter_wise.append(avg_accuracy_for_k_fold)
        

    print("\n***************** RESULTS *****************")
    for p in m:
        print('\navg_accuracy for smoothing parameter m = %s:'%p,avg_accuracy_parameter_wise[p])
    #print(avg_accuracy_parameter_wise)



    figure(figsize=(12, 6), dpi=80)
    for i in m:
        plt.plot(k,avg_accuracy_parameter_wise[i])
    #plt.plot(k,avg_accuracy_parameter_wise[0],'r')
    #plt.plot(k,avg_accuracy_parameter_wise[1],'b')
    plt.xlabel('K_Fold')
    plt.ylabel('Accuracy')
    plt.legend(m)
    print("\n\n************* Plot of Accuracy **************\n")
    plt.show()
    
    
    standard_deviation_for_parameters = []
    for i in range(len(avg_accuracy_parameter_wise)):
        for j in range(len(avg_accuracy_parameter_wise[i])):     
            x=abs(avg_accuracy_parameter_wise[i][j] - np.mean(avg_accuracy_parameter_wise[i]))**2
        standard_deviation_for_parameters = standard_deviation_for_parameters+ [np.sqrt(np.mean(x))]
        #standard_deviation_for_parameters.append[np.sqrt(np.mean(x))]
    print('\nstandard deviations :',standard_deviation_for_parameters)


    from matplotlib.pyplot import figure
    figure(figsize=(12, 6), dpi=80)
    m = [0,1]
    plt.plot(m,standard_deviation_for_parameters)
    plt.xlabel('m_Parameters')
    plt.ylabel('Standard Deviation of Accuracies')
    plt.legend(m)
    print("\n\n************* Plot of Standard Deviation **************")
    plt.show()
    
end = time.time()
print('\n Time taken to execute : %s seconds'%(end - start))
```
