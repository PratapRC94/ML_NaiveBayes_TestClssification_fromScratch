## For m=0 and m=1 smoothing parameter

### - plot averages of the accuracy and standard deviations (as error bars) as a function of train set size.

```python
# Import the libraries
import re
import os
import sys
import math
import random
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import figure

import time
start = time.time() #starting a clock time to keep track of the run time

# function to read the files residing in a given directory
def read_files(directory):
    data = {}
    for files in os.listdir(directory):
        with open(os.getcwd() + "/" + directory + "/" + files, "r",encoding='Latin-1') as f:
            data[files] = f.read()
    return data

# function to parse through each text and cleaning the text for our model
def pre_processing(text):
    
    # Convert string to lower case
    text1 = str(text).lower() 
    # Removing html tags and content inside it
    text1 = re.sub(r'<.*?>', '', text1)
    # Removing URLs
    text1 = re.sub(r'[a-z]*[:.]+\S+', '', text1)
    # Remove punctuations
    text1 = text1.replace("'","")
    text1 = re.sub(r'[^\w\s]',' ',text1)
    # Remove hyphen,underscore
    text1 = re.sub('-|_', ' ',text1)
    # Removing digits
    text1 = re.sub('\d+', "", text1)
    # Removing words with 2 letters and less
    text1 = re.sub(r'\b\w{1,2}\b', '',text1)
    # Remove extra space with ''
    text1 = re.sub("\s\s+"," ",text1)
    
    return text1

# Main function
if __name__ == "__main__": 

    data = read_files('pp1data') # pass the directory name as it is a standard folder name for the data
    # to find all the available text data files and asks for user input choice
    files = list(data.keys())
    for i in range(len(files)):
        print("Enter %d for %s"%((i+1),files[i]))
    inp = files[int(input())-1]
    print("\n*********** Processing the data file : %s ***********"%inp)
    source_data = data.get(inp)
    
    # text data processing - split based on \n and \t 
    am = source_data.lower().split("\n")
    # removing any empty string from the row of the data
    am = list(string for string in am if len(string)>0)
    
    # splitting based on \t and store the text and class label as a dictionary value
    am_processed={}
    for i in range(len(am)):
        am_processed[i] = list([pre_processing(am[i].split('\t')[0]),am[i].split('\t')[1]])
    len(am_processed)

    # initialise the given smoothing parameter 'm' and fold 'k'
    m = [0,1]
    k=10
    avg_accuracy_parameter_wise = [] # list to store final avg accuracy for each m per training set size
    sd_parameter_wise = [] # list to store final standard deviation of accuracy for each m per training set size
    
    # iterate for each m
    for p in m :
        train_allfolds_accuracy = [] # list of average of accuracies over k_fold
        # iterate for k folds
        for j in range(k):
            # random shuffle the dataset as its sequancial and equally distributed to class 0 and 1
            am_to_be_shuffled = list(am_processed.items())
            random.shuffle(am_to_be_shuffled)
            am_shuffled = dict(am_to_be_shuffled)

            #creating test data set from entire vocabulary - dictionary contains both text and label
            test = dict(list(am_shuffled.items())[j*int(len(am_shuffled)/k):(j*int(len(am_shuffled)/k))+int(len(am_shuffled)/k)])
            #creating train data set from entire vocabulary
            train = {}
            for i in set(am_shuffled.keys()) - set(test.keys()):
                train[i] = am_shuffled[i]
            
            #splitting the train and test data according to class label 0 and 1 with only text data in dictionary
            lb_1_test={} #pos
            lb_1_train={}
            lb_0_test={} #neg
            lb_0_train={}

            for i in (test.keys()):
                if test[i][1]=='0':
                    lb_0_test[i] = test[i][0]
                else:
                    lb_1_test[i] = test[i][0]

            # fixing the test set, iterate through k folds for k sub-samples of training set
            train_points=[]
            train_fold_accuracy=[]
            for q in range(k):
                accuracy= []
                # taking training sets for each sub-samples 
                train_folds = dict(list(train.items())[0:(q*int(len(train)/k))+int(len(train)/k)])
                train_points.append(len(train_folds))
                for i in (train_folds.keys()):
                    if train_folds[i][1]=='0':
                        lb_0_train[i] = train_folds[i][0]
                    else:
                        lb_1_train[i] = train_folds[i][0]
                #print(len(lb_1_train),len(lb_0_train))

                #****************************************************************************************************************

                #merging class 0/1 wise train and test data
                training_data = {**lb_1_train,**lb_0_train}
                test_data = {**lb_1_test,**lb_0_test}
                
                #storing all the words from documents of class 0 and class 1
                token_1 = []
                for sent in lb_1_train.values():
                    token_1.extend(sent.split())

                token_0 = []
                for sent in lb_0_train.values():
                    token_0.extend(sent.split())
                # creating the training vocabulary
                train_tokens = token_1+token_0
                #****************************************************************************************************************
                # finding the probability of class 1 and class 0 in the selected train set
                
                #prob_1 = np.log(len(lb_1_train)/(len(lb_1_train) + len(lb_0_train)))
                #prob_0 = np.log(len(lb_0_train)/(len(lb_1_train) + len(lb_0_train)))

                prob_1 = len(lb_1_train)/(len(lb_1_train) + len(lb_0_train))
                prob_0 = len(lb_0_train)/(len(lb_1_train) + len(lb_0_train))
                #print(prob_1,prob_0)
                #****************************************************************************************************************

                # for tokens in train_tokens but not in class 0
                train_tokens_not_0 = set(train_tokens) - set(token_0)
                train_tokens_not_0_dict = dict.fromkeys(list(train_tokens_not_0), 0)
                # dictionary of frequency of each token in class 0
                token_0_mle = dict(Counter(token_0))
                # all tokens and frequency count for class 0
                neg_counts = {**train_tokens_not_0_dict,**token_0_mle}
                #MAP estimate of tokens of class 0
                map_neg_prob ={}
                for key in neg_counts.keys():
                    #map_neg_prob[key] = np.log((neg_counts[key]+p)/(sum(token_0_mle.values())+(p*len(set(train_tokens)))))
                    map_neg_prob[key] = (neg_counts[key]+p)/(sum(token_0_mle.values())+(p*len(set(train_tokens))))

                # for tokens in train_tokens but not in class 1
                train_tokens_not_1 = set(train_tokens) - set(token_1)
                train_tokens_not_1_dict = dict.fromkeys(list(train_tokens_not_1), 0)
                # dictionary of frequency of each token in class 1
                token_1_mle = dict(Counter(token_1))
                # all tokens and frequency count for class 1
                pos_counts = {**train_tokens_not_1_dict,**token_1_mle}
                #MAP estimate of tokens of class 1
                map_pos_prob ={}
                for key in pos_counts.keys():
                    #map_pos_prob[key] = np.log((pos_counts[key]+p)/(sum(token_1_mle.values())+(p*len(set(train_tokens)))))
                    map_pos_prob[key] = (pos_counts[key]+p)/(sum(token_1_mle.values())+(p*len(set(train_tokens))))

                #****************************************************************************************************************
                ## class prediction on test data set

                final_class = {}
                for key,sent in test_data.items():    
                    words = sent.split()
                    # skip the words not available in training set
                    available_words = set(words).intersection(set(train_tokens)) 

                    #MAP estimate of tokens from test data using the MAP of tokens from train data 
                    class_1 = 1
                    class_0 = 1

                    #class_1 += prob_1
                    #class_0 += prob_0
                    # MAP estimate for tokens in document from test set
                    for word in available_words:
                        #class_0 = class_0 + np.log(map_neg_prob[word])
                        #class_1 = class_1 + np.log(map_pos_prob[word])
                        class_0 = class_0 * map_neg_prob[word]
                        class_1 = class_1 * map_pos_prob[word]
                    class_0 = class_0 * prob_0
                    class_1 = class_1 * prob_1
                    
                    # Class 0 and 1 probability comparison and classification of class
                    if class_1 > class_0:
                        final_class[key] = '1'
                    elif class_1 == class_0:
                        temp = random.randint(0, 1)
                        if temp == 1:
                            final_class[key] = '1'
                        else:
                            final_class[key] = '0'
                    else:
                        final_class[key] = '0'
                #****************************************************************************************************************
                # Original class labels for each documents in test set
                actual_test_class = {}
                for key in test.keys():
                    actual_test_class[key] = test[key][1]
                #****************************************************************************************************************
                # Accuracy
                accuracy.append(len(set(final_class.items()).intersection(set(actual_test_class.items())))/len(actual_test_class))
                #****************************************************************************************************************
                # taking average of accuracy of all the test document and store as accuracy for one training sub-sample
                avg_of_accuracy = np.mean(np.array(accuracy))
                # store the accuracy from each training sub-sample for each of the k-fold
                train_fold_accuracy.append(avg_of_accuracy)
                
            # create an accuracy matrix for k-fold and each fold has accuracy for each training sub-samples
            train_allfolds_accuracy.append(train_fold_accuracy)
            
            # finding avg of accuracy and standard deviation per training set size
            narr = np.array(train_allfolds_accuracy)
            avg_accuracy_for_k_fold = list(np.mean(narr, axis=0))
            sd_for_k_fold = list(np.std(narr,axis=0))


        # store the avg of accuracy and standard deviations for each smoothing parameter (m) experiment
        avg_accuracy_parameter_wise.append(avg_accuracy_for_k_fold)
        sd_parameter_wise.append(sd_for_k_fold)

    # print the results and plots

    print("\n***************** RESULTS *****************")
    for p in m:
        print('\navg_accuracy for smoothing parameter m = %s:'%p,avg_accuracy_parameter_wise[p])
        print('\nSD_ for smoothing parameter m = %s:'%p,sd_parameter_wise[p])
    
    print('\n training data points :',train_points)

    figure(figsize=(12, 6), dpi=80)
    for i in m:
        plt.errorbar(train_points, avg_accuracy_parameter_wise[i],yerr=sd_parameter_wise[i],fmt='-o')

    plt.xlabel('Training set size')
    plt.ylabel('Avg accuracy per training set')
    plt.title('************* Plot of Accuracy : %s **************'%inp)
    plt.legend(m)
    print('\n************* Plot of Accuracy **************\n')
    plt.show()
    
    figure(figsize=(12, 6), dpi=80)
    for i in m:
        plt.errorbar(train_points, sd_parameter_wise[i],yerr=sd_parameter_wise[i],fmt='-o')

    plt.xlabel('Training Set Size')
    plt.ylabel('Standard deviation per training set')
    plt.title('************* Plot of Standard Deviation : %s **************'%inp)
    plt.legend(m)
    print('\n************* Plot of Standard Deviation **************\n')
    plt.show()

# ending the clock to get the total execution time of the function
end = time.time()
print('\n Time taken to execute : %s seconds'%(end - start))
```
