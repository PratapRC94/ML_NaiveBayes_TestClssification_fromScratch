```python
def read_files(directory):
    data = {}
    for files in os.listdir(directory):
        with open(os.getcwd() + "/" + directory + "/" + files, "r",encoding='Latin-1') as f:
            #data[files] = pre_processing(f.read()) 
            data[files] = f.read()
    return data
```

```python
def pre_processing(text):
    
    # Convert string to lower case
    text1 = str(text).lower()
    
    # Replacing \n
    #text1 = str(text1).replace("\n", " ")
    
    # Replacing \t
    #text1 = str(text1).replace("\t", " ")
    
    # Removing html tags and content inside it
    text1 = re.sub(r'<.*?>', '', text1)
    
    # Removing URLs
    text1 = re.sub(r'[a-z]*[:.]+\S+', '', text1)
    
    # Remove punctuations
    text1 = text1.replace("'","")
    text1 = re.sub(r'[^\w\s]',' ',text1)
    text1 = text1.replace("  "," ")
    
    # Remove hyphen,underscore
    text1 = re.sub('-|_', '',text1)
    
    # Removing digits
    text1 = re.sub("\d+", "", text1)
    
    # Removing words with 2 letters and less
    text1 = re.sub(r'\b\w{1,2}\b', '',text1)
    
    
    return text1
```

```python
def words(doc1):
    """
    Function to Split the document to words
    
    Parameters:
    doc1 - documents to be split
    
    Returns:
    List of words for each document
    
    """
    return[word for sent in doc1 for word in sent.split()]





def word_count_maxl(words):
    """
    Function to calculate maximum likelihood of a word given that class
    
    Parameters:
    words_count - word counts of a class
    words_all   - All the words in that class
    
    Returns:
    Maximum likelihood of a word given the class
    
    """
    words_mle = dict(Counter(words))
    
    # Removing sparse words
    #words_mle = dict(filter(lambda elem: elem[1] > 10, words_mle.items()))
    
    total_words = sum(words_mle.values())
    
    words_mle_final = {}
    for words in list(words_mle.keys()):
        words_mle_final[words] = np.log(words_mle[words]/total_words)
        
    
    return words_mle_final
```

```python
m = [0,1]
k=[2,3,4,5,6,7,8,9,10]
avg_accuracy_parameter_wise = [] # list of average of accuracies over k_fold(2,3,4,5,6,7,8,9,10) over m(0,1)
for p in m :
    avg_accuracy_for_k_fold = [] # list of average of accuracies over k_fold(2,3,4,5,6,7,8,9,10)
    for q in k:
        accuracy= [] # list of accuracy for each iteration of k_fold
        for j in range(q):
            #creating test data set from entire vocabulary - dictionary contains both text and label
            test = dict(list(am_processed.items())[j*int(len(am_processed)/q):(j*int(len(am_processed)/q))+int(len(am_processed)/q)])
            #creating train data set from entire vocabulary
            train = {}
            for i in set(am_processed.keys()) - set(test.keys()):
                train[i] = am_processed[i]

            #splitting the train and test data according to class label 0 and 1 with only text data in dictionary
            lb_1_test={} #pos
            lb_1_train={}
            lb_0_test={} #neg
            lb_0_train={}

            for i in (train.keys()):
                if train[i][1]=='0':
                    lb_0_train[i] = train[i][0]
                else:
                    lb_1_train[i] = train[i][0]

            for i in (test.keys()):
                if test[i][1]=='0':
                    lb_0_test[i] = test[i][0]
                else:
                    lb_1_test[i] = test[i][0]

            #print(len(lb_1_test),len(lb_0_test))
            #****************************************************************************************************************

            #merging class 0/1 wise train and test data
            training_data = {**lb_1_train,**lb_0_train}
            test_data = {**lb_1_test,**lb_0_test}

            #token_1 = words(list(lb_1_train.values()))
            #token_0 = words(list(lb_0_train.values()))
            token_1 = []
            for sent in lb_1_train.values():
                for word in sent.split():
                    token_1.append(word)
                    
            token_0 = []
            for sent in lb_0_train.values():
                for word in sent.split():
                    token_0.append(word)

            train_tokens = token_1+token_0
            #****************************************************************************************************************

            prob_1 = np.log(len(lb_1_train)/(len(lb_1_train) + len(lb_0_train)))
            prob_0 = np.log(len(lb_0_train)/(len(lb_1_train) + len(lb_0_train)))
            #print(prob_1,prob_0)
            #****************************************************************************************************************

            # for tokens in train_tokens but not in class 0
            train_tokens_not_0 = set(train_tokens) - set(token_0)
            train_tokens_not_0_dict = dict.fromkeys(list(train_tokens_not_0), 0)
            #dictionary of frequency of each token in class 0
            token_0_mle = dict(Counter(token_0))

            neg_counts = {**train_tokens_not_0_dict,**token_0_mle}

            #MAP estimate of tokens of class 0
            map_neg_prob ={}
            for key in neg_counts.keys():
                map_neg_prob[key] = np.log((neg_counts[key]+p)/(sum(token_0_mle.values())+(p*len(set(train_tokens)))))


            # for tokens in train_tokens but not in class 1
            train_tokens_not_1 = set(train_tokens) - set(token_1)
            train_tokens_not_1_dict = dict.fromkeys(list(train_tokens_not_1), 0)
            #dictionary of frequency of each token in class 1
            token_1_mle = dict(Counter(token_1))

            pos_counts = {**train_tokens_not_1_dict,**token_1_mle}
            #MAP estimate of tokens of class 1
            map_pos_prob ={}
            for key in pos_counts.keys():
                map_pos_prob[key] = np.log((pos_counts[key]+p)/(sum(token_1_mle.values())+(p*len(set(train_tokens)))))

            #****************************************************************************************************************
            ## class prediction on test data set

            final_class = {}
            for key,sent in test_data.items():    
                words = sent.split()
                available_words = set(words).intersection(set(train_tokens)) #skip the words not available in training set

                #MAP estimate of tokens from test data using the MAP of tokens from train data 
                class_1 = 0
                class_0 = 0

                class_1 += prob_1
                class_0 += prob_0

                for word in available_words:
                    class_0 = class_0 + np.log(map_neg_prob[word])
                    class_1 = class_1 + np.log(map_pos_prob[word])

                if class_1 > class_0:
                    final_class[key] = '1'
                elif class_1 == class_0:
                    temp = random.randint(0, 1)
                    if temp == 1:
                        final_class[key] = '1'
                    else:
                        final_class[key] = '0'
                else:
                    final_class[key] = '0'
            #****************************************************************************************************************
            actual_test_class = {}
            for key in test.keys():
                actual_test_class[key] = test[key][1]
            #****************************************************************************************************************
            # Accuracy
            accuracy.append(len(set(final_class.items()).intersection(set(actual_test_class.items())))/len(actual_test_class))
        sum_of_accuracy = 0
        for i in range(len(accuracy)):
            sum_of_accuracy = sum_of_accuracy + accuracy[i]
    
        avg_accuracy = sum_of_accuracy/len(accuracy)
       # print(avg_accuracy)
        avg_accuracy_for_k_fold = avg_accuracy_for_k_fold + [avg_accuracy]

   # print(avg_accuracy_for_k_fold)
    avg_accuracy_parameter_wise = avg_accuracy_parameter_wise + [avg_accuracy_for_k_fold]

print(avg_accuracy_parameter_wise)
```
